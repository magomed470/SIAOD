Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
class Stack:
    def __init__(self):
        self.items = []
        
    def __len__(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []

    def append(self, key):
        self.items.append(key)

    def pop(self):
        return self.items.pop()

    def peek(self):
        """ Проверяет, но не удаляет элемент, находящийся в верхней части стэка """
        return self.items[-1]


class Deque:
    def __init__(self):
        self.items = []
        
    def __len__(self):
        return len(self.items)

    def is_empty(self):
        return self.items == []

    def append(self, key):
        self.items.append(key)

    def append_left(self, key):
        self.items.insert(0, key)

    def pop(self):
        return self.items.pop()

    def pop_left(self):
        return self.items.pop(0)

    def peek(self):
        return self.items[-1]

    def peek_left(self):
        return self.items[0]
```

### Задание №1
Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.

```python
filename = 'booksforlab4.txt'

def sorted_books():
    with open(filename, 'r', encoding='utf-8') as books:
        titles = Deque()
        sorted_titles = Deque()

        for book in books:
            print(book.rstrip())
            titles.append(book)
        print()
        
        while not titles.is_empty():
            arr = titles.pop()

            while not sorted_titles.is_empty() and sorted_titles.peek() < arr:
                titles.append_left(sorted_titles.pop())
            sorted_titles.append(arr)

        while not sorted_titles.is_empty():
            print(sorted_titles.pop().rstrip())


sorted_books()
```

### Задание №2
Дек содержит последовательность символов для шифровки сообщений. Дан текстовый файл, содержащий зашифрованное сообщение. Пользуясь деком, расшифровать текст. 
Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.

```python
filename = 'encodedmessageforlab4.txt'

alphabet = 'abcdefghijklmnopqrstuvwxyz'
print(alphabet,'\n')


keyRing = Deque()
for letter in alphabet:
    keyRing.append(letter)


def encode_char(c):
    for i in range(len(keyRing)):
        x = keyRing.pop_left()
        if x == c:
            keyRing.append(x)
            val = keyRing.pop_left()
            keyRing.append(val)
            return val
        keyRing.append(x)


def decode_char(c):
    for i in range(len(keyRing)):
        x = keyRing.pop()
        if x == c:
            keyRing.append_left(x)
            val = keyRing.pop()
            keyRing.append_left(val)
            return val
        keyRing.append_left(x)


def encoded_msg():
    with open(filename, 'r', encoding='utf-8') as text:
        encoded = ''
        for line in text:
            if encoded_letter := encode_char(line):
                encoded += encoded_letter
            else:
                encoded += line
        print(encoded)

        decoded = ''
        for line in encoded:
            if decoded_letter := decode_char(line):
                decoded += decoded_letter
            else:
                decoded += line
        print(decoded)


encoded_msg()
```

### Задание №3
Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
- на каждом шаге со стержня на стержень переносить только один диск;
- диск нельзя помещать на диск меньшего размера;
- для промежуточного хранения можно использовать стержень В.
Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
def hanoi(disks, A, C):
    if disks <= 0:
        return None
    B = 6 - A - C         # Промежуточный стержень; 6 - плюсуем номера стержней (1+2+3)
    hanoi(disks - 1, A, B)
    print('Перенести диск', disks, 'со стержня', A, 'на стержень', C)
    hanoi(disks - 1, B, C)

hanoi(4, 1, 3)
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
def check_brackets(string):
    bracket_stack = Stack()

    for symbol in string:
        if symbol == '(':
            bracket_stack.append(symbol)
        elif symbol == ')':

            if bracket_stack.is_empty():
                return False
            bracket_stack.pop()
    return bracket_stack.is_empty()


print(check_brackets('()(((( ))))()'))
print(check_brackets('()(((( ))))())'))
```

### Задание №5
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс квадратных скобок в тексте, используя дек.

```python
def check_square_brackets(string):
    bracket_deque = Deque()

    for symbol in string:
        if symbol == '[':
            bracket_deque.append(symbol)
        elif symbol == ']':

            if bracket_deque.is_empty():
                return False
            bracket_deque.pop()
    return bracket_deque.is_empty()

print(check_square_brackets('[][][][[[[ ]]]]]'))
print(check_square_brackets('[][][][[ ]]'))
```

### Задание №6
Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.

```python
def stack_symbols():
    text = 'Александр Сергеевич Пушкин родился 6 июня 1799 года в Москве'

    letters = Stack()
    numbers = Stack()
    others = Stack()

    for line in text:
        for element in line:
            if element.isalpha():
                letters.append(element)
            elif element.isdigit():
                numbers.append(element)
            else:
                others.append(element)

    new_text = ''

    while not others.is_empty():
        new_text += others.pop()

    while not letters.is_empty():
        new_text += letters.pop()

    while not numbers.is_empty():
        new_text += numbers.pop()

    print(new_text[::-1])


stack_symbols()
```

### Задание №7
Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, затем все положительные числа, сохраняя исходный порядок в каждой группе.

```python
filename = 'numsforlab4.txt'

def deque_nums():
    with open(filename, 'r', encoding='utf-8') as nums:
        deque = Deque()
        
        for line in nums:
            print(line, '\n')
            temp = line.split(' ')

            for element in temp:
                if int(element) < 0:
                    deque.append_left(element)
                else:
                    deque.append(element)

        while not deque.is_empty():
            value = deque.pop_left()

            if int(value) < 0:
                deque.append(value)
            else:
                deque.append_left(value)
                break

        while not deque.is_empty():
            value = deque.pop()

            if int(value) < 0:
                print(value)
            else:
                deque.append(value)
                break

        while not deque.is_empty():
            print(deque.pop_left())


deque_nums()
```

### Задание №8
Дан текстовый файл. Используя стек, сформировать новый текстовый файл, содержащий строки исходного файла, записанные в обратном порядке: первая строка становится последней, вторая – предпоследней и т.д.

```python
filename = 'stihforlab4.txt'


def stack_rev():
    with open(filename, 'r', encoding='utf-8') as poem:
        stack = Stack()

        for line in poem:
            line = line.strip()
            print(line)
            stack.append(line)
        print()

        while not stack.is_empty():
            print(stack.pop())


stack_rev()
```
